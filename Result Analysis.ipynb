```python
aggregated_data = pd.read_excel('AAAI_analysis.xlsx')
```

```python
aggregated_data = aggregated_data.groupby(by=["Score Type","GPT Model", "Prompt Type"], dropna=False).sum()
```

```python
import matplotlib.pyplot as plt

# Filter data for GPT3.5 and GPT4
gpt35_data = aggregated_data.xs('GPT3.5', level='GPT Model')
gpt4_data = aggregated_data.xs('GPT4', level='GPT Model')

# Unstacking the data for plotting
gpt35_unstacked = gpt35_data['Frequency'].unstack(level='Score Type').fillna(0)
gpt4_unstacked = gpt4_data['Frequency'].unstack(level='Score Type').fillna(0)

# Setting up the plots
fig, axes = plt.subplots(1, 2, figsize=(15, 6))

# Plotting for GPT3.5
gpt35_unstacked.plot(kind='bar', stacked=True, ax=axes[0])
axes[0].set_title('GPT3.5')
axes[0].set_xlabel('Prompt Type')
axes[0].set_ylabel('Frequency')

# Plotting for GPT4
gpt4_unstacked.plot(kind='bar', stacked=True, ax=axes[1])
axes[1].set_title('GPT4')
axes[1].set_xlabel('Prompt Type')
axes[1].set_ylabel('Frequency')

# Adding the legend and showing the plot
plt.legend(title='Score Type', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
# Customizing the plot with different colors, horizontal x-axis labels, and Times New Roman font

# Setting the font to Times New Roman
plt.rcParams['font.family'] = 'Times New Roman'

# Custom colors for the stacked bar chart
colors = ['#F5F5F5', '#DAE8FC', '#D5E8D4', '#FFE6CC', '#F8CECC', '#FFF2CC']
# (-1-1, 0-0, 0-1, 0.5-0, 1-0, 1-1)

# Setting up the plots again
fig, axes = plt.subplots(1, 2, figsize=(15, 6))

# Plotting for GPT3.5 with customizations
gpt35_unstacked.plot(kind='bar', stacked=True, color=colors, ax=axes[0])
axes[0].set_title('GPT3.5')
axes[0].set_xlabel('Prompt Type')
axes[0].set_ylabel('Frequency')
axes[0].tick_params(axis='x', rotation=0)  # Making x-axis labels horizontal

# Plotting for GPT4 with customizations
gpt4_unstacked.plot(kind='bar', stacked=True, color=colors, ax=axes[1])
axes[1].set_title('GPT4')
axes[1].set_xlabel('Prompt Type')
axes[1].set_ylabel('Frequency')
axes[1].tick_params(axis='x', rotation=0)  # Making x-axis labels horizontal

#horizontal grid only 
axes[1].grid(axis='y', linestyle='dashed', linewidth=0.5)
axes[0].grid(axis='y', linestyle='dashed', linewidth=0.5)

#legent at bottom
# Put a legend below current axis
axes[0].legend(loc='upper center', bbox_to_anchor=(0.5, -0.05),fancybox=True, ncol=5)
axes[1].get_legend().set_visible(False)
axes[0].get_legend().set_visible(False)

# Adjusting the legend and layout
#plt.legend(title='Score Type', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()

plt.savefig('result.png')
plt.show()

```
